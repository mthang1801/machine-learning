## Aggregate

```python
matrix = np.random.randint(0,10000,1000)
```

```python
np.min(matrix)
```

```python
np.max(matrix)
```

```python
np.mean(matrix)
```

```python
np.median(matrix)
```

```python
np.var(matrix)
```

```python
np.std(matrix)
```

```python
np.sqrt(matrix)
```

```python
# Standardize 
mean = np.mean(matrix)
std = np.std(matrix)
matrix_scale = ((matrix - mean)/std).round(2)
```

```python
matrix_scale
```

```python
%matplotlib inline
import matplotlib.pyplot as plt 
plt.hist(matrix_scale)
plt.show()
```

## Dot product example

```python
np.random.seed(0)
sales_amount = np.random.randint(20,size=(5,3))
sales_amount
```

```python
# Weekly sale data frame
weekly_sales = pd.DataFrame(sales_amount, index=["Mon","Tue", "Wed", "Thu", "Fri"],
                            columns=["Almond butter", "Peanut butter", "Cashew butter"])
```

```python
weekly_sales
```

```python
# Create prices array 
prices = np.array([10,8,12])
prices
```

```python
# Create  butter price DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3),
                             columns=["Almond butter", "Peanut butter", "Cashew butter"], 
                             index=["Price"])
butter_prices
```

```python
# total_amount = np.dot(weekly_sales, prices.reshape(3,1))
weekly_sales["Total ($)"] =  weekly_sales.dot(prices.reshape(3,1))
```

```python
weekly_sales
```

```python
pd.concat([weekly_sales,butter_prices],axis=0)
```
